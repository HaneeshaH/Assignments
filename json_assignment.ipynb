```python

import requests
import pandas as pd
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp
```

```python
data=resp.json()
data
```

```python
df=pd.DataFrame(data)
df1 = df['user'].apply(pd.Series)
df1.head()
df1.columns
```

```python
df2 = df['labels'].apply(pd.Series)
df2=df2.add_prefix('labels_')
```

```python
final_df = pd.concat([df1,df2],axis=1)
final_df
#len(final_df)
```
