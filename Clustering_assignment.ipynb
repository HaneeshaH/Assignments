### Problem Statement
In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.The raw image is available in spicy.misc package with the name face.


```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Here, we are going to do color segmentation. We will simply assign pixels to the same segment if they have a similar color.

```python
image=misc.face(gray=True)
```

```python
image.shape
```

The image is represented as a 2D array: the first dimensionâ€™s size is the height, the second is the width.

 The following code reshapes the array to get a long list of gray values, then it clusters these gray-values using K-Means.

```python
X=image.reshape(-1,1)
```

```python
X.shape
```

Let's choose k=5

```python
#Clustering
kmeans1=KMeans(n_clusters=5,random_state=100)
kmeans1.fit(X)
```

```python
segmented_img1 = kmeans1.cluster_centers_[kmeans1.labels_]
```

```python
segmented_img1.shape
```

```python
segmented_img1 = segmented_img1.reshape(image.shape)
```

```python
segmented_img1.shape
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
axs[0].set_title('Original image')
axs[0].imshow(image,cmap='gray')
axs[1].set_title('Compressed image(K=5)')
axs[1].imshow(segmented_img1,cmap='gray')
```
